# YouTube Data API (Search oldest videos on a topic)

```python
# We need to set the time according to google api
from datetime import datetime
start_time = datetime(year=2005, month=1, day=1).strftime('%Y-%m-%dT%H:%M:%SZ')
end_time = datetime(year=2008, month=1, day=1).strftime('%Y-%m-%dT%H:%M:%SZ')
```

```python
request = youtube.search().list(part='snippet',
                            q='python programming',
                            type='video',
                            publishedAfter=start_time,
                            publishedBefore=end_time,
                            maxResults=50)
```

```python
result = request.execute()
```

```python
# Let's sort the video on the bases of published date in ascending order
sorted(result['items'], key=lambda x:x['snippet']['publishedAt'])
```

```python
for item in sorted(result['items'], key=lambda x:x['snippet']['publishedAt']):
    print(item['snippet']['title'], item['snippet']['publishedAt'])
```

## YouTube Data API (Get all videos of a channel)

```python
def get_channel_videos(channel_id):
    
    # get Uploads playlist id
    res = youtube.channels().list(id=channel_id, 
                                  part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    next_page_token = None
    
    while 1:
        res = youtube.playlistItems().list(playlistId=playlist_id, 
                                           part='snippet', 
                                           maxResults=50,
                                           pageToken=next_page_token).execute()
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
    
    return videos
```
